```python
#drop rows not containing data
sd_crime = sd_crime.drop(labels = [0,1,2,76,77,78], axis = 0)
```

```python
#change column names and reset index
sd_crime.columns = ['year', 'population', 'murder', 'rape', 'robbery', 'aggravated assault', 'violent crime', 'burglary', 'larceny', 'vehicle theft', 'property crime', 'index total']
sd_crime = sd_crime.reset_index(drop = True)
sd_crime
```
